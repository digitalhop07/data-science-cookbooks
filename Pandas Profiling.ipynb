# Pandas Profiling Cookbook

## Why use it?
[Pandas Profiling](https://github.com/pandas-profiling/pandas-profiling) is a Python library that provides a quick solution for calculating descriptive statistics from a Pandas DataFrame.

## How to install:
- Add `pandas_profiling` to the `requirements.txt` file in your ScienceBox repo and then run `sb --build`

## Example usage:

```python
import pandas_profiling
```

```python
# Use Chartify to generate some example data
import chartify
data = chartify.examples.example_data()
```

```python
# Aggregate the data
aggregated_data = (data.groupby(['country', 'fruit'])
                   .agg({'total_price': 'sum', 'unit_price': 'mean', 'quantity': 'sum'}))
```

```python
aggregated_data
```

## Note: The HTML output below looks pretty bad in GHE, but is rendered nicely when you open the notebook in Jupyter.

```python
pandas_profiling.ProfileReport(aggregated_data)
```

```python
# Aggregate by time
timeseries_data = (data.groupby(['date', 'fruit'])
                   .agg({'total_price': 'sum', 'unit_price': 'mean', 'quantity': 'sum'}))
```

```python
pandas_profiling.ProfileReport(timeseries_data)
```
